Importing The Dependencies........!!

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataframe
credit_card_data = pd.read_csv('/content/creditcard_data.csv')
```

```python
# Distribution of Fraud Transcation & Legit transaction..!
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0 --> Legit Transaction
1--> Fraudlent Transaction

```python
# Separating The data for analysis
Legit = credit_card_data[credit_card_data.Class == 0]
Fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
 # Staticsical measure of the data
Legit.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
# Compare Values For both transaction
credit_card_data.groupby('Class').mean()
```

Under-Sampling the data

Build a sample dataset containg similar distribution of Legit Transcation & Fraudlent Transaction.

The number of Fraudlent transaction is 492

```python
Legit_sample = Legit.sample(n=492)
```

concatenating Two dataframes..!

```python
new_dataset = pd.concat([Legit_sample,Fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

Spiliting the data into Target & Features

```python
X = new_dataset.drop(columns='Class' , axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split The data into Training Data & Testing Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
# Trainng The logistic regression model with traing data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
 # Accuracy on training data..!
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, classification_report

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1-score:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
